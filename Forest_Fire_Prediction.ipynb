## **Importing Data**

## **Data Preprocessing**

```python
df['month'] = df['month'].replace({'jan':1,'feb':2,'mar':3,'apr':4,'may':5,'jun':6,'jul':7,'aug':8,'sep':9,'oct':10,'nov':11,'dec':12})
df['day'] = df['day'].replace({'sun':1,'mon':2,'tue':3,'wed':4,'thu':5,'fri':6,'sat':7})
df
```

```python
df.columns
```

```python
dataset = df[['month','day','FFMC','DMC','DC','ISI','temp','RH','wind','rain','area','size_category']]
```

```python
dataset = dataset.rename({'size_category':'Size'},axis = 1)
dataset
```

```python
le = LabelEncoder()
dataset['Size'] = le.fit_transform(dataset['Size'])
```

```python
df
```

## **Model Building**

```python
x = dataset.drop(labels=['area'], axis=1)
y = dataset[["area"]]
```

```python
mms = MinMaxScaler()
scaled_x = mms.fit_transform(x)
```

```python
x_train, x_test, y_train, y_test = train_test_split(scaled_x, y, test_size=0.2, random_state=19)
```

```python
x_train.shape, x_test.shape, y_train.shape, y_test.shape
```

```python
model = Sequential([
    Dense(64, activation='relu', input_shape=(11,)),
    Dense(64, activation='relu'),
    #Dense(32, activation='relu'),
    Dense(32, activation='relu'),
    Dense(1, activation='sigmoid'),
])
```

```python
model.compile(optimizer='sgd',
              loss='categorical_crossentropy',
              metrics=['accuracy'])
```

```python
hist = model.fit(x_train, y_train,
          batch_size=10, epochs=20,
          validation_data=(x_test,y_test))
```

## **Model Validation**

```python
model.evaluate(x_test, y_test)[1]
```
